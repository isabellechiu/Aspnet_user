# Top 10 Common Words for id: 205783 

```python
import pandas as pd
q = pd.read_csv("C:/Users/DCUK/id205783.csv", usecols=["question"])
q = q["question"].tolist()
q = " ".join(str(i) for i in q)
q = q.lower().split()
q1 = []
for i in q:
    if i not in stop_words:
       q1.append(i)
q2 = []
for word in q1:
    word = word.replace(".", "")
    word = word.replace(",","")
    word = word.replace(":","")
    word = word.replace("\"","")
    word = word.replace("!","")
    word = word.replace("â€œ","")
    word = word.replace("â€˜","")
    word = word.replace("*","")
    q2.append(word)
```
